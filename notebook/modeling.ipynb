# 1. 데이터 정리

```python
import pandas as pd
df = pd.read_csv('../data/swim_eda2.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.dropna()
df.info()
```

```python
print(f'buy_size: {df["buy_size"].unique()}')
print(f'size_diff: {df["size_diff"].unique()}')
```

```python
sorted(df['usual_size'].unique()),sorted(df['buy_size'].unique())
```

```python
# size remapping
size_map = { '2XS':0, 'XS': 1, 'S': 2, 'M': 3, 'L': 4, 'XL': 5, '2XL': 6, '3XL': 7}
df.loc[:,'size_num'] = df['buy_size'].map(size_map)
df.loc[:,'usual_size_num'] = df['usual_size'].map(size_map)
```

```python
sorted(df['usual_size_num'].unique()),sorted(df['size_num'].unique())
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
# 성별 추가
def extract_gender(name):
    if '남자' in name:
        return '남성'
    elif '여자' in name:
        return '여성'
    else:
        return '기타'  # 해당 없음
    
df['gender'] = df['product_name'].fillna('').apply(extract_gender)
```

```python
df['gender'].value_counts()
```

```python
df = df[df['gender']!='기타']
df['gender'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
mms = MinMaxScaler()
le = LabelEncoder()
df['gender_le']= le.fit_transform(df['gender'])
df['height_mms']= mms.fit_transform(df[['height']])
df['weight_mms'] = mms.fit_transform(df[['weight']])
df.head()
```

```python
# extract brand name
df['brand'] = df['product_name'].str.split().str[0]
df.head()
```

```python
df['brand'].value_counts()
```

```python
df.columns
```

```python
# 상위 6개 브랜드만 따로 저장
top6_brand = ['스웨이브','나이키','후그','센티','키치피치','르망고']
df2 = df[df['brand'].isin(top6_brand)]

# 브랜드명 컬럼 추가
brand_df = pd.concat([df2,pd.get_dummies(df2['brand'])],axis=1)
brand_df = brand_df.reset_index(drop=True)
brand_df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,f1_score,classification_report
from sklearn.preprocessing import MinMaxScaler
from imblearn.over_sampling import SMOTE
from sklearn.svm import SVC
from lightgbm import LGBMClassifier
  


X = brand_df[['height_mms','weight_mms','usual_size_num','gender_le']]
y = brand_df['buy_size']

sm = SMOTE()
X_resampled,y_resampled = sm.fit_resample(X,y)

X_train,X_test,y_train,y_test = train_test_split(X_resampled,y_resampled,test_size=0.2,random_state=42)

lgbm = LGBMClassifier(random_state=42)
lgbm.fit(X_train,y_train)

y_pred_tr_lgbm = lgbm.predict(X_train)
y_pred_ts_lgbm = lgbm.predict(X_test)

svc = SVC(kernel='linear')
svc.fit(X_train,y_train)

y_pred_tr_svc = svc.predict(X_train)
y_pred_ts_svc = svc.predict(X_test)

print(f'train_acc(lgbm):',accuracy_score(y_train,y_pred_tr_lgbm))
print(f'test_acc:(lgbm)',accuracy_score(y_test,y_pred_ts_lgbm))
print(classification_report(y_test, y_pred_ts_lgbm, zero_division=0))

print(f'train_acc(SVC):',accuracy_score(y_train,y_pred_tr_svc))
print(f'test_acc(SVC):',accuracy_score(y_test,y_pred_ts_svc))
print(classification_report(y_test, y_pred_ts_svc, zero_division=0))
```

## LGBM & SMOTE 사용
* hyperparameter 조정을 통한 성능 향상 시도하기

```python
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score, classification_report
from imblearn.over_sampling import SMOTE
from lightgbm import LGBMClassifier
import pandas as pd
import time
import json

X = brand_df[['height_mms', 'weight_mms', 'usual_size_num', 'gender_le'] + top6_brand]
y = brand_df['buy_size']

sm = SMOTE(random_state=42)
X_resampled, y_resampled = sm.fit_resample(X, y)

X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

param_grid = {
    'num_leaves': [31, 50],
    'max_depth': [-1, 10],
    'learning_rate': [0.05, 0.1],
    'n_estimators': [100, 300],
    'min_child_samples': [20, 50],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}

lgbm = LGBMClassifier(random_state=42)

log_file = "grid_search_log.txt"
with open(log_file, "w") as log:
    log.write("Grid Search Start...\n")

def log_progress(msg):
    with open(log_file, "a") as log:
        log.write(f"{time.strftime('%Y-%m-%d %H:%M:%S')} - {msg}\n")

grid_search = GridSearchCV(
    estimator=lgbm,
    param_grid=param_grid,
    scoring='accuracy',
    cv=3,              # cv=5로 했을 때 결과나오는 시간이 너무 오래 걸려서 줄였습니다!
    n_jobs=-1,
    verbose=1
)

start_time = time.time()
grid_search.fit(X_train, y_train)
end_time = time.time()

log_progress(f"Grid Search Completed. Time Taken: {end_time - start_time:.2f} seconds")

best_params = grid_search.best_params_
best_score = grid_search.best_score_
log_progress(f"Best Parameters: {json.dumps(best_params)}")
log_progress(f"Best CV Score: {best_score:.4f}")

best_lgbm = grid_search.best_estimator_
y_pred_tr = best_lgbm.predict(X_train)
y_pred_ts = best_lgbm.predict(X_test)

train_acc = accuracy_score(y_train, y_pred_tr)
test_acc = accuracy_score(y_test, y_pred_ts)

log_progress(f'Train Accuracy: {train_acc:.4f}')
log_progress(f'Test Accuracy: {test_acc:.4f}')

classification_rep = classification_report(y_test, y_pred_ts, zero_division=0)
log_progress(f"\nClassification Report:\n{classification_rep}")

results_df = pd.DataFrame(grid_search.cv_results_)
results_df = results_df[['params', 'mean_test_score', 'rank_test_score']].sort_values(by='mean_test_score', ascending=False)

log_progress("\n상위 5개 하이퍼파라미터 조합:")
for idx, row in results_df.head(5).iterrows():
    log_progress(f"Params: {row['params']} | Mean Test Score: {row['mean_test_score']:.4f} | Rank: {row['rank_test_score']}")

print("\n상위 5개 하이퍼파라미터 조합:")
print(results_df.head(5))
```

```python
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score, classification_report
from imblearn.over_sampling import SMOTE
from lightgbm import LGBMClassifier
import pandas as pd
import time
import json

# 데이터 준비
X = brand_df[['height_mms', 'weight_mms', 'usual_size_num', 'gender_le'] + top6_brand]
y = brand_df['buy_size']

# SMOTE로 데이터 균형 조정
sm = SMOTE(random_state=42)
X_resampled, y_resampled = sm.fit_resample(X, y)

# Train/Test 분리
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)


lgbm = LGBMClassifier(random_state=42, colsample_bytree=0.8, learning_rate=0.1, max_depth=10, min_child_samples=50, n_estimators=300, num_leaves=31, subsample= 0.8)
lgbm.fit(X_train,y_train)

y_pred_tr = lgbm.predict(X_train)
y_pred_ts = lgbm.predict(X_test)

print(f'train:',accuracy_score(y_train,y_pred_tr))
print(f'test:',accuracy_score(y_test,y_pred_ts))
```

```python
# 최적의 파라미터로 새 모델 생성
best_params = grid_search.best_params_
best_params['random_state'] = 42  # random_state는 따로 추가

best_lgbm = LGBMClassifier(**best_params)
best_lgbm.fit(X_train, y_train)

# 예측 및 평가
y_pred_tr = best_lgbm.predict(X_train)
y_pred_ts = best_lgbm.predict(X_test)

train_acc = accuracy_score(y_train, y_pred_tr)
test_acc = accuracy_score(y_test, y_pred_ts)

print(f"Train Accuracy with Best Params: {train_acc:.4f}")
print(f"Test Accuracy with Best Params: {test_acc:.4f}")

# 상세 평가 리포트
print("\nClassification Report on Test Set:")
print(classification_report(y_test, y_pred_ts, zero_division=0))
```
